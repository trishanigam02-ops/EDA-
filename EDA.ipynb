#Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

- 1. What is Overfitting?
Meaning:

Overfitting happens when a model learns too much from the training data ‚Äî including noise and random fluctuations.
It performs very well on training data but poorly on new/test data.

How to Detect Overfitting:

High training accuracy, but low test/validation accuracy

Training loss keeps decreasing, but validation loss increases

Large performance gap between training and validation

Model is unnecessarily complex for the dataset

How to Prevent Overfitting:

Cross-Validation (e.g., k-fold CV)
Helps test model performance on multiple data splits.

Regularization Techniques

L1 (Lasso)

L2 (Ridge)
Penalize large weights to simplify the model.

Dropout (used in Deep Learning)
Randomly drops neurons during training to avoid memorization.

Early Stopping
Stop training when validation loss stops improving.

Simplify the Model
Use fewer layers, fewer parameters, or simpler algorithms.

Collect More Data
More data reduces noise and improves generalization.

Feature Selection
Remove irrelevant features that cause the model to memorize patterns.

2. What is Underfitting?
Meaning:

Underfitting happens when a model is too simple and cannot learn the underlying pattern.
Both training and test accuracy remain low.

How to Detect Underfitting:

Low training performance

Low test performance

High training and validation error

Model fails to capture trends even in training data

How to Prevent Underfitting:

Increase Model Complexity
Use more layers, deeper trees, or more powerful algorithms.

Train for More Epochs
Give the model more time to learn patterns.

Reduce Regularization Strength
Too much regularization can oversimplify the model.

Add More Features
Introduce relevant variables to improve learning.

Use Better Algorithms
Move from simple models (like linear regression) to advanced ones (like Random Forest or Gradient Boosting).

3. Bias‚ÄìVariance Tradeoff (In Simple Words)

Overfitting = Low bias + High variance
Model is extremely flexible and fits training data too well.

Underfitting = High bias + Low variance
Model is rigid and cannot learn enough.

Goal: Find the right balance to achieve the best accuracy on unseen data.

#Question 4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).
- 1. Oversampling (SMOTE)
Theory:

Oversampling increases the number of samples in the minority class so the dataset becomes balanced.
SMOTE (Synthetic Minority Oversampling Technique) creates synthetic (artificial but realistic) samples instead of simply duplicating existing ones.

How SMOTE works:

It finds nearest neighbors of a minority sample

Creates new artificial data points between them

Helps the model learn better decision boundaries

Practical Example:

Suppose you have:

Class 0: 10,000 samples

Class 1: 900 samples

After applying SMOTE:

Class 1 increases to ~10,000 synthetic samples

Dataset becomes balanced

Model becomes better at detecting Class 1

Code Example (Python):


```python
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification

# 1. Create a sample dataset
X, y = make_classification(n_samples=1000, n_features=5,
                           weights=[0.9, 0.1], random_state=42)

# 2. Apply SMOTE
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

print("Original:", X.shape, "Resampled:", X_resampled.shape)
```

2. Random Undersampling / Oversampling
A) Random Oversampling
Theory:

Increase the minority class by duplicating existing samples.

Practical Example:

Fraud class = 1,000
Duplicate samples until it becomes 10,000.

Pros: Simple
Cons: Risk of overfitting because data is repeated.
B) Random Undersampling
Theory:

Reduce the size of the majority class by removing samples.

Practical Example:

Instead of 99,000 non-fraud samples, randomly keep only 5,000.
Dataset becomes balanced.

Pros: Fast, less memory
Cons: You may lose important information.
3. Class Weights in Models (Bonus Technique)
Theory:

You tell the model to penalize misclassifying the minority class more.
This way, the model gives extra importance to rare cases.

Practical Example:

In Logistic Regression:

```python
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler

# Create dataset
X, y = make_classification(n_samples=500, n_features=4,
                           weights=[0.8, 0.2], random_state=0)

# Apply oversampling
ros = RandomOverSampler(random_state=0)
X_res, y_res = ros.fit_resample(X, y)
```

The algorithm automatically increases weight of the minority class.

Use Cases:

Fraud detection

Medical diagnosis

Spam detection

Accident prediction

#Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and STANDARIZATION
-Feature scaling means bringing all numerical features to a similar range.
It is important because many ML algorithms behave poorly when features have very different scales.

Why Feature Scaling is Important
1. Distance-Based Algorithms Perform Better

Algorithms like:

KNN

K-Means

SVM

use distance calculations (Euclidean distance).
If one feature has a large range (e.g., Salary 0‚Äì1,00,000)
and another has a small range (e.g., Age 18‚Äì60),

then the large-range feature dominates the distance calculation.

Scaling fixes this by making all features contribute equally.

2. Gradient Descent Converges Faster

Algorithms like:

Linear Regression

Logistic Regression

Neural Networks

use gradient descent.

If features have different scales, gradient descent takes longer, zig-zags, and may not converge properly.

Scaling makes gradients smooth and helps faster, stable learning.

3. Improves Model Accuracy

Models become:

More stable

Less variance

Better optimization

More consistent learning

#Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
-Categorical variables in a dataset must be converted into numbers before training a machine-learning model. Two common techniques are Label Encoding and One-Hot Encoding.

1. Label Encoding
What it does:

Converts each category into a unique integer.

Example:
Color = {Red, Blue, Green}
Label Encoded ‚Üí {Red=0, Blue=1, Green=2}

When to Use It:

‚úî When the categorical variable is ordinal (has a natural order)

Examples of ordinal data:

Size: Small < Medium < Large

Education level: High school < Bachelor < Master < PhD

Customer rating: 1 Star < 2 Stars < ‚Ä¶ < 5 Stars

Pros:

Simple

Creates no extra columns

Cons:

Creates false numeric order if applied to nominal data

Example: If Red=0 and Green=2, the model may think ‚ÄúGreen > Red,‚Äù which is incorrect.

2. One-Hot Encoding
What it does:

Creates separate binary columns for each category.

Example:
Color = Red, Blue, Green

One-Hot Encoding ‚Üí

Red ‚Üí [1, 0, 0]

Blue ‚Üí [0, 1, 0]

Green ‚Üí [0, 0, 1]

When to Use It:

‚úî When the categorical variable is nominal (NO natural order)

Examples of nominal data:

City (Delhi, Mumbai, Kolkata)

Gender (Male, Female, Other)

Product type (Mobile, Laptop, TV)

Pros:

No false order introduced

Ideal for nominal categories

Cons:

Creates many new columns (high dimensionality)

Not suitable when categories are too many (e.g., thousands of cities)

To understand how the category of an app affects its rating, we calculate the average rating for each category.

#Question 8: Titanic Dataset
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
survival rate, and why do you think that happened?
b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
adults (Age ‚â• 18). Did children have a better chance of survival?
Dataset: https://github.com/MasteriNeuron/datasets.git




```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset (Titanic CSV) from correct raw link
# Correct raw link for Titanic dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv"
df = pd.read_csv(url)

# Step 3: Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()

# Step 4: Check first 5 rows
df.head()
```

```python
# Step 5: Survival rate by passenger class (Pclass)
pclass_survival = df.groupby('Pclass')['Survived'].mean()
print("Survival rate by Pclass:\n", pclass_survival)

# Optional: Bar plot
plt.figure(figsize=(6,4))
sns.barplot(x=pclass_survival.index, y=pclass_survival.values, palette='viridis')
plt.title("Survival Rate by Passenger Class (Pclass)")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Step 6: Create Age group column: Children (<18) vs Adults (>=18)
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

# Step 7: Survival rate by AgeGroup
agegroup_survival = df.groupby('AgeGroup')['Survived'].mean()
print("\nSurvival rate by AgeGroup:\n", agegroup_survival)

# Optional: Bar plot
plt.figure(figsize=(6,4))
sns.barplot(x=agegroup_survival.index, y=agegroup_survival.values, palette='coolwarm')
plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.show()
```

#Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.
b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?
Dataset: https://github.com/MasteriNeuron/datasets.git


```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset from raw GitHub CSV link
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

# Step 3: Clean column names
df.columns = df.columns.str.strip()

# Drop the 'Unnamed: 0' column if it exists, as it's often an artifact from saving DataFrames
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Create a 'route' column by combining source and destination cities
df['route'] = df['source_city'] + '-' + df['destination_city']

# Step 4: Ensure numeric columns and handle missing data
df['price'] = pd.to_numeric(df['price'], errors='coerce')        # Convert price to numeric
df['days_left'] = pd.to_numeric(df['days_left'], errors='coerce')  # Convert days_left to numeric
df = df.dropna(subset=['price', 'days_left', 'airline', 'route']) # Drop rows with missing key values

# Step 5: Check first few rows
df.head()

# Scatter plot for price vs days_left
plt.figure(figsize=(10,6))
sns.scatterplot(x='days_left', y='price', data=df)
plt.title("Flight Prices vs Days Left Until Departure")
plt.xlabel("Days Left to Departure")
plt.ylabel("Flight Price")
plt.show()

# Average price by days_left
mean_price_by_days = df.groupby('days_left', as_index=False)['price'].mean()
plt.figure(figsize=(10,6))
sns.lineplot(x='days_left', y='price', data=mean_price_by_days, marker='o')
plt.title("Average Flight Price vs Days Left")
plt.xlabel("Days Left to Departure")
plt.ylabel("Average Price")
plt.show()

# Filter for route Delhi-Mumbai
route_df = df[df['route'].str.strip() == 'Delhi-Mumbai']

# Average price per airline
airline_price = route_df.groupby('airline', as_index=False)['price'].mean().sort_values(by='price')
print("Average Price per Airline (Delhi-Mumbai):\n", airline_price)

# Bar plot visualization
plt.figure(figsize=(10,6))
sns.barplot(x='airline', y='price', data=airline_price, palette='Set2', hue='airline', legend=False)
plt.title("Average Flight Price by Airline (Delhi-Mumbai)")
plt.xlabel("Airline")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```
